```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def antennaModel(beam_angle:float) -> float:
    base = np.cos(np.deg2rad(beam_angle)) ** (2.99999)
    return base
```

```python
phi = np.arange(0, 360, 1)
```

```python
pattern = antennaModel(phi)
```

```python
plt.polar(np.rad2deg(phi), pattern)
```
