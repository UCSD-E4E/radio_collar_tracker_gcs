```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def antennaModel(beam_angle:float) -> float:
    base = np.cos(beam_angle) ** (2.99999)
    return base
```

```python
phi = np.arange(0, 2 * np.pi, 0.01)
```

```python
pattern = antennaModel(phi)
```

```python
plt.polar(phi, pattern)
```

```python
my_heading = 45
my_pos = np.array([1, 1, 0])
tx_pos = np.array([2, 1, 0])
```

```python
# Get my current heading
x = np.cos(np.deg2rad(my_heading))
y = np.sin(np.deg2rad(my_heading))
heading_vector = np.array([x, y, 0])
heading_vector
```

```python
# Get the bearing to the transmitter
transmitter_vector = tx_pos - my_pos
transmitter_vector = transmitter_vector / np.linalg.norm(transmitter_vector)
transmitter_vector
```

```python
# Compute the bearing from antenna to transmitter
dot = np.dot(heading_vector, transmitter_vector)
angle = np.arccos(dot)
np.rad2deg(angle)
```

```python
antennaModel(angle)
```
